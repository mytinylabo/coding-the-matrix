```python
# テキストファイルは http://resources.codingthematrix.com/ から取得する
f = open('stories_small.txt')
stories = list(f)
len(stories)
```

```python
# p44 課題 0.6.6
def make_inverse_index(strlist):
    wordlist = set(' '.join(strlist).split())
    docs = list(enumerate(strlist))
    return {w: set(map(lambda doc: doc[0], filter(lambda doc: w in doc[1], docs))) for w in wordlist}
```

```python
%time make_inverse_index(stories)
```

```python
# list(enumerate...) したくなかった版
from collections import defaultdict
from itertools import tee
def make_inverse_index2(strlist):
    wordlist = set(' '.join(strlist).split())
    rindex = defaultdict(set)
    docs = enumerate(strlist)
    for w in wordlist:
        docs, working_docs = tee(docs)
        rindex[w] = set(map(lambda doc: doc[0], filter(lambda doc: w in doc[1], working_docs)))
    return rindex
```

```python
# 速度ほぼ変わらず…
%time make_inverse_index2(stories)
```

```python
# p44 課題 0.6.7
from functools import reduce
def or_search(inverse_index, query):
    return reduce(lambda a, b: a | b, map(lambda q: inverse_index[q] if q in inverse_index else set(), query))
```

```python
rindex = make_inverse_index(stories)
```

```python
or_search(rindex, ['Olympic', 'Sesame', 'chorus', 'fifth', 'hoge'])
```

```python
# p44 課題 0.6.8
def and_search(inverse_index, query):
    return reduce(lambda a, b: a & b, map(lambda q: inverse_index[q] if q in inverse_index else set(), query))
```

```python
and_search(rindex, ['Olympic', 'Sesame', 'chorus', 'fifth', 'hoge'])
```

```python
and_search(rindex, 'A prep course for the month-long World Cup'.split())
```

```python
# ちょっと戻って p16 暗号の話
bad_table = {
    (0, 'C'): 0,
    (0, 'H'): 1,
    (0, 'S'): 1,
    (1, 'C'): 1,
    (1, 'H'): 0,
    (1, 'S'): 0
}

def encode(plain, key, table):
    return table[(plain, key)]  

def decode(cipher, key, table):
    return [p for (p, k), c in table.items() if cipher == c and key == k][0]
```

```python
# 監視者であるイブは、通信メッセージが上記のテーブルで暗号化されていることを知っている。
# ただしどのキーが使われているかは分からない。
# イブは暗号化されたメッセージを見て、元のメッセージについての情報を得ることができるだろうか？

# イブが見たメッセージは 1 だとする。
# どのキーが使われるかは一様分布だが、それで復号した平文の確率分布はどうなるか？
import random
result = defaultdict(int)
for i in range(0, 1000):
    result[ decode(1, ['C','H','S'][random.randint(0, 2)], bad_table) ] += 1
result
```

```python
# メッセージが 1 の場合、平文が 0 である確率の方が 1 である確率の 2 倍になる。
```

```python
# p44 課題 0.8.1
def tuple_sum(a, b):
    return [(a1 + b1, a2 + b2) for (a1, a2), (b1, b2) in zip(a, b)]

tuple_sum([(1,2), (10,20)], [(3,4), (30,40)])
```
