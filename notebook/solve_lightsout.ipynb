```python
def make_rule_mat(dim):
    cols = []
    for i in range(dim**2):
        v = Vec(set(range(dim**2)), {})
        x = i % dim
        y = i // dim
        for dx, dy in [(-1, 0), (1, 0), (0, -1), (0, 1), (0, 0)]:
            xi = x + dx
            yi = y + dy
            if (0 <= xi < dim) and (0 <= yi < dim):
                v[yi * dim + xi] = one
        cols.append(v)
    return coldict2mat(cols)

def pp_mat(mat):
    mat.pp(sorted(mat.D[0]), sorted(mat.D[1]))

def pp_vec(vec):
    n = int(math.sqrt(len(vec.D)))
    for i in range(len(vec.D)):
        print('■ ' if vec[i] == one else '□ ', end='')
        if i % n == n - 1:
            print()
```

```python
def get_problem_vec():
    url = 'http://127.0.0.1:8080/problem/latest'
    problem = requests.get(url).json()
    
    return Vec(set(range(problem['dimension']**2)), {k: one for k in problem['on_cells']})
```

```python
A = make_rule_mat(3)
pp_mat(A)
```

```python
b = get_problem_vec()
pp_vec(b)
```

```python
x = solve(A, b)
pp_vec(x)
```
